```python
# part 2
'''
Technically this is the best possible result here. It has the lowest error-bars 
and does not involved the Helmholtz to Gibbs correction. All enthalpies here 
also include the Jacobean of the flexible triangular box. 

    In other words, there are no assumptions here related to the protocol.
        Ergodicity of MD data is the only remaining assumption overall, 
        but these simulations were ran for quite a long time already.

However, the amount of data and number evaluation (in each Form) compared to 400K result earlier are:

## ## ## ## 

# Tref = 400K, NVT PGM:  [using 8 temperature replicas in MBAR]

    # FF evaluations MBAR       : 5600000                  # 5.6 M
    # FF evaluations BAR (PGM)  : (20000/50)*5000          # 2  M
    
    # ns MD NVT (for PGM)       : (1000000 / 10000) ns     # 100ns
    # ns MD NPT (used in MBAR)  : 8 * (100000 / 10000) ns  # 80ns
    
# Tref = 200K, NPT PGM:  [using 10 temperature replicas in MBAR]

    # FF evaluations MBAR       : 9000000                  # 9 M
    # FF evaluations BAR (PGM)  : the same                 # 2 M
    
    # ns MD NPT (for PGM)       : (800000 / 10000) ns      # 80ns
    # ns MD NPT (used in MBAR)  : 10 * (100000 / 10000) ns # 100ns (90 ns if taking the data from the above step instead)

## ## ## ## 

Totals including all 3 Forms (I, II, III):

# Tref = 400K, NVT PGM: 
    # FF evaluations : (5.6 + 2) * 3  # 22.8 M
    # ns MD          : (100 + 80) * 3 # 540  ns

# Tref = 200K, NPT PGM:
    # FF evaluations : (9 + 2) * 3    # 33 M
    # ns MD          : (80 + 100) * 3 # 540  ns

## ## 

Thus, the most precise result took 10.2 M more energy evaluations compared to the 2nd most precise.
    This difference turns out only from having more states inside MBAR.

In conclusion:
    For this system it was useful to connect the second Tx ~ 375K (I and II) to the first Tx ~ 200K (II and III)
    Therefore the MBAR overheads of having 10 instead of 8 states was worthwhile.
    Therefore (Tref = 200K, NPT PGM; the final figure of the notebook) seems to be the main result.
    
    Ways to improve the overall cost:
        Try MBAR with less data at each temperature replica (10ns is quite a lot at the moment).
            This is done below using 5ns from each replicate,
                but the problem is that this data subset is taken from the back each trajectory.
                It is therefore less likely to be ergodic compared to 10ns.
                The data needs to be strided/shuffled first before taking a smaller subset.
                    This aspect is still a TODO inside Tx.py

# copy-pase from the later cell: might be the best res
    # FF evaluations : (4.5 + 2) * 3  # 19.5 M
    # ns MD          : (80 + 50) * 3  # 390  ns
'''
```
