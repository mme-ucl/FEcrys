## All data (100,000 samples / state for MBAR)

```python
key_min = 'III'

# part 2
ax = plot_curves_(curves_npt, y_lim=[-2,14], # [-2,14]
                  y_lim_enthalpy=[-4,8], #y_lim=[-4,12], y_lim_enthalpy=[-2,2],
                  units_kJ_per_mol=True,
                  x_lim=[100,650],
                  loc_Forms='upper left', loc_ax0='upper center', loc_ax1='upper center',
                  subplots_args= [1,2],
                 subplots_kwargs= {'dpi': 300, 'figsize':(8,4)},
                  key_min=key_min,
                  capsize=3,
                  y_stride=2,
                 title_size = 12,
                 label_size = 12, 
                 )
ax = plot_curves_under_(curves_npt_1ns, ind_min=2, ax=ax) ## grey dotsb!

n_mol = 16
for form in list_Forms:
    for T in [200,300,400]:
        delta_f     = (nns_nvt[(form,T,n_mol)].nns[0].BAR_V_FE - nns_nvt[(key_min,T,n_mol)].nns[0].BAR_V_FE) / n_mol
        delta_f_se  = (nns_nvt[(form,T,n_mol)].nns[0].BAR_V_SE + nns_nvt[(key_min,T,n_mol)].nns[0].BAR_V_SE) / n_mol
        delta_f    *= (CONST_kB*T)
        delta_f_se *= (CONST_kB*T)
        
        #ax[0].errorbar([T], [delta_f], [delta_f_se], color='red', fmt="", capsize=5, zorder=100)
        ax[0].scatter([T], [delta_f], s=10, color='red', zorder=1000)
```
