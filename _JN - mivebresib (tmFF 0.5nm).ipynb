```python
# part 2 

# adding the missing data between 200 and 300K for MBAR
NPT_(
_list_Forms = ['I','II','III'],
_Ts = [225, 275],
_list_n_mol = {'I':[16],'II':[16],'III':[16],},
NPT_subfolder = NPT_subfolder,
save_equilibrated_supercell = False,

checking = True,
checking_check_only_path = False,
checking_try_plotting_equlibrated_cell = True, # will not work because not saved at these Ts
#checking_resave_datasets_with_CUT_xyz = True, # already done becuase this was fast to load
)
```

```python
# part 2
n_mol_Tref = 16
n_mol_NPT = 16
        
Tref = 200

''' each Form cost:
mbar : number of FF evaluations involved: 9000000 : 9M
''';
curves_npt = {}
for Form in ['I', 'II','III']:
    curves_npt[Form] = MBAR_(
                    Form = Form,
                    Tref_FEref_SEref = [Tref,
                                        nns_npt[(Form, Tref, n_mol_Tref)].nns[0].BAR_V_FE / n_mol_Tref, # absolute_lattice_FE
                                        nns_npt[(Form, Tref, n_mol_Tref)].nns[0].BAR_V_SE / n_mol_Tref, # absolute_lattice_SE
                                        ], # notebook 
                    FEref_SEref_is_Helmholtz = False,
                    list_Ts = [200, 225, 250, 275, 300, 350, 400, 450, 500, 550],
                    batch_size = 100000,
                    n_mol_NPT = n_mol_NPT,
                    NPT_subfolder = NPT_subfolder,
                    clear_memory = True,
                    xyz_not_in_datasets = True,
                    )
    
''' each Form cost:
mbar : number of FF evaluations involved: 900000  : 0.9M
''';
curves_npt_1ns = {}
for Form in ['I', 'II','III']:
    curves_npt_1ns[Form] = MBAR_(
                    Form = Form,
                    Tref_FEref_SEref = [Tref,
                                        nns_npt[(Form, Tref, n_mol_Tref)].nns[0].BAR_V_FE / n_mol_Tref, # absolute_lattice_FE
                                        nns_npt[(Form, Tref, n_mol_Tref)].nns[0].BAR_V_SE / n_mol_Tref, # absolute_lattice_SE
                                        ], # notebook 
                    FEref_SEref_is_Helmholtz = False,
                    list_Ts = [200, 225, 250, 275, 300, 350, 400, 450, 500, 550],
        
                    batch_size = 10000,  use_representative_subsets=True, # representative here means to say: as ergodic as possible
                                        ## 'randomly drawn from each temperature replica' 
                                        ## x[random_indices][:10], rather than x[-10:], for getting 10 elemnts from x.
        
                    n_mol_NPT = n_mol_NPT,
                    NPT_subfolder = NPT_subfolder,
                    clear_memory = True,
                    xyz_not_in_datasets = True,
                    )
```
